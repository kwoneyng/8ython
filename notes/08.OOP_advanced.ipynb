## ì‹¤ìŠµ - ìŠ¤íƒœí‹±(ì •ì ) ë©”ì„œë“œ

> ê³„ì‚°ê¸° classì¸ `Calculator`ë¥¼ ë§Œë“¤ì–´ë´…ì‹œë‹¤.

* ë‹¤ìŒê³¼ ê°™ì´ ì •ì  ë©”ì„œë“œë¥¼ êµ¬ì„±í•œë‹¤. 
* ëª¨ë“  ì •ì  ë©”ì„œë“œëŠ”, ë‘ ìˆ˜ë¥¼ ë°›ì•„ì„œ ê°ê°ì˜ ì—°ì‚°ì„ í•œ ê²°ê³¼ë¥¼ ë¦¬í„´í•œë‹¤.
* `a` ì—°ì‚°ì `b` ì˜ ìˆœì„œë¡œ ì—°ì‚°í•œë‹¤. (`a - b`, `a / b`)
    1. `add(a, b)` : ë§ì…ˆ

    3. `mul(a, b)` : ê³±ì…ˆ
    4. `div(a, b)` : ë‚˜ëˆ—ì…ˆ

````python
# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class Calculator:
    
    @staticmethod
    def add(a, b):
        return a+b
    
    @staticmethod
    def sub(a, b):
        return a-b
    
    @staticmethod
    def mul(a, b):
        return a*b
    
    @staticmethod
    def div(a, b):
        if b != 0:
            return a / b
        else :
            return '0ìœ¼ë¡  ë‚˜ëˆŒ ìˆ˜ ì—†ì–´'
````

````python
# ì •ì  ë©”ì„œë“œë¥¼ í˜¸ì¶œí•´ë³´ì„¸ìš”.
print(Calculator.add(1, 5), Calculator.sub(3, 2), Calculator.mul(3, 7))
mal = Calculator()
mal.div(1,5)
````

````python
# ì‚¬ëŒê³¼ ì‚¬ëŒì„ ê°™ì€ì§€ ë¹„êµí•˜ë©´, ì´ëŠ” ë‚˜ì´ê°€ ê°™ì€ì§€ ë¹„êµí•œ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ë„ë¡ ë§Œë“¤ì–´ë´…ì‹œë‹¤.
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __repr__(self):
        return self.age
    
    def __gt__(self, other):
        if self.age > other.age :
            return True
        else :
            return False
        
    def __lt__(self, other):
        if self.age < other.age :
            return True
        else :
            return False
````

## í¬ì¼“ëª¬ êµ¬í˜„í•˜ê¸°

ìš°ì„  ì•„ë˜ì™€ ê°™ì´ êµ¬í˜„í•´ ë³´ê³ , ì¶”ê°€ë¡œ ë³¸ì¸ì´ ì›í•˜ëŠ” ëŒ€ë¡œ êµ¬í˜„ ë° ìˆ˜ì •í•´ ë´…ì‹œë‹¤.

* `exp`: ê²½í—˜ì¹˜
    * ìƒëŒ€ë°©ì„ ì“°ëŸ¬ëœ¨ë¦¬ë©´ ìƒëŒ€ë°© `level` * 15 ë¥¼ íšë“í•©ë‹ˆë‹¤.
    * ê²½í—˜ì¹˜ëŠ” `level` * 100 ì´ ë˜ë©´, ë ˆë²¨ì´ í•˜ë‚˜ ì˜¬ë¼ê°€ê³  0ë¶€í„° ì¶”ê°€ ë©ë‹ˆë‹¤. 

ì´í›„ ì´ìƒí•´ì”¨, íŒŒì´ë¦¬, ê¼¬ë¶€ê¸°ëŠ” í¬ì¼“ëª¬ì„ ìƒì†í•˜ì—¬ ììœ ë¡­ê²Œ êµ¬í˜„í•´ ë´…ì‹œë‹¤.

* í¬ì¼“ëª¬ => ë¬¼í¬ì¼“ëª¬ => ê¼¬ë¶€ê¸° 
* í¬ì¼“ëª¬ => ë¬¼í¬ì¼“ëª¬ => ì‰ì–´í‚¹
* í¬ì¼“ëª¬ => ë¹„í–‰í¬ì¼“ëª¬, ë¶ˆí¬ì¼“ëª¬ => íŒŒì´ì–´

ì™€ ê°™ì´ ë‹¤ì–‘í•œ ì¶”ê°€ ìƒì†ê´€ê³„ë„ êµ¬í˜„í•´ ë´…ì‹œë‹¤.

````python
import random

# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class Pokemon:
    
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * 20
        self.exp = 0
        
    
    def bark(self):
        print(f'{self.name}ì˜ ìš¸ë¶€ì§–ê¸°!')
        print('ê·¸ëŸ¬ë‚˜ ì•„ë¬´ ì¼ë„ ì¼ì–´ë‚˜ì§€ ì•Šì•˜ë‹¤.. ')
        

    def check_battle(self, target):
        if target.hp <= 0:
            print(f'\n{target.name}ì´(ê°€) ì“°ëŸ¬ì¡ŒìŠµë‹ˆë‹¤! ğŸ’« ')
            self.exp += target.level * 15
            
            print(f'{self.name}ì˜ ìŠ¹ë¦¬!')
            
            if self.exp >= self.level * 100:
                self.exp -= self.level * 100
                self.level += 1
                self.hp = self.level * 20
                print(f'ë ˆë²¨ì—…ì„ í•˜ì—¬ {self.level} ì´ ë˜ì—ˆê³ , í˜„ì¬ ê²½í—˜ì¹˜ëŠ” {self.exp} ì…ë‹ˆë‹¤!')
                
                
    def battle(self, target):
        count = 0
        while self.hp >= 0 and target.hp >= 0:
            print(f'#### ë¼ìš´ë“œ {count} ####')
            print(f'{self.name}ì˜ ì²´ë ¥ : {self.hp}')
            print(f'{target.name}ì˜ ì²´ë ¥ : {target.hp}')
            print('-----------------------------------')
            
            attacker = random.choice([self, target])
            defencer = target if attacker == self else self
        
            skill_number = random.randint(1, 3)
            
            if skill_number == 1:
                attacker.property_skill(defencer)
            elif skill_number == 2:
                attacker.character_skill(defencer)
            else:
                attacker.bark()
            
            count += 1
            print('\n')
        
        self.hp = self.level * 20
        target.hp = target.level * 20
````

````python
class WaterPokemon(Pokemon):
    property = 'water'
    
    def property_skill(self, target):
        damage = self.level * 3
        target.hp -= damage
        print(f'{self.name}ì˜ ë¬¼ëŒ€í¬! ğŸ’¦ ')
        print(f'{target.name}ì—ê²Œ {damage} ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤!')
        self.check_battle(target)
        
class ThunderPokemon(Pokemon):
    property = 'thunder'
    
    def property_skill(self, target):
        damage = self.level * 4
        target.hp -= damage
        print(f'{self.name}ì˜ ê°ì „ì‹œí‚¤ê¸°')  
        print(f'{target.name}ì—ê²Œ {damage} ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤.')
        self.check_battle(target)
        
class NormalPokemon(Pokemon):
    property = 'normal'
    
    def property_skill(self, target):
        damage = self.level * 4
        target.hp -= damage
        print(f'{self.name}ì˜ ëª¸í†µë°•ì¹˜ê¸°!!! ğŸ‘ŠğŸ»ğŸ‘ŠğŸ»')  
        print(f'{target.name}ì—ê²Œ {damage} ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…í˜”ë‹¤.')
        self.check_battle(target)
````
